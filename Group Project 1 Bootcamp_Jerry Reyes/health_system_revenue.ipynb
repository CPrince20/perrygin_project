```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('investment.csv')
data['Revenues of financing schemes'].unique()
```

```python
data
```

```python
data1 = data.loc[data.Year == 2015]
```

```python
data1.Measure.unique()
```

```python
data2 = data1.loc[data1.Measure == 'Share of financing scheme']
data3 = data2.loc[data2['Financing scheme'] == 'All financing schemes']
data3 = data3.sort_values('Country').reset_index(drop=True)
```

```python
data3
```

```python
labelsCanada = ['Transfer from Goverment', 'Social Insurance', 'Voluntary Prepayment', 'Other Domestic Revenue']
colors_canada = ['r','brown', 'tomato', 'lightsalmon']

plt.pie(data3.loc[data3.Country=='Canada'].Value,\
        labels = labelsCanada, colors = colors_canada,\
        autopct="%1.1f%%", shadow=True, startangle=0, explode=[0.2, 0.1, 0.1, 0.1])
plt.title('CANADA', fontsize=14)
plt.savefig('Canada_revenue.png')
```

```python
labels_usa = ['Transfer from Goverment', 'Social Insurance','Compulsory Prepayment',\
                'Voluntary Prepayment', 'Other Domestic Revenue']
colors_usa = ['blue','darkslateblue','slateblue','mediumpurple','mediumslateblue']
plt.pie(data3.loc[data3.Country=='United States'].Value,\
        labels = labels_usa, colors = colors_usa,\
        autopct="%1.1f%%", shadow=True, startangle=0, explode=[0.2, 0.1, 0.1, 0.1, 0.1])
plt.title('UNITED STATES', fontsize = 14)
plt.savefig('united_states_revenue.png')
```

```python
data4 = data[data.Measure == 'Share of financing scheme']
data5 = data4[data4['Financing scheme'] == 'All financing schemes']
data6 = data5[data5['Revenues of financing schemes'] == 'Transfers from government domestic revenue']
data6.reset_index(drop=True)
```

```python
plt.plot(data6.loc[data6.LOCATION=='CAN'].TIME, data6.loc[data6.LOCATION=='CAN'].Value)
```

```python
len(data6.loc[data6.LOCATION=='USA'].Value)
```

```python
plt.plot(data6.loc[data6.LOCATION=='USA'].TIME, data6.loc[data6.LOCATION=='USA'].Value)
```

```python
ax1 = plt.subplot(121, ylabel = "Revenue's percent from Goverments Transfer")           # Default margin is 0.05, value 0 means fit
ax1.plot(data6.loc[data6.LOCATION=='CAN'].TIME, data6.loc[data6.LOCATION=='CAN'].Value, color='r')
ax1.set_title('CANADA')
ax2 = plt.subplot(122)      # Values >0.0 zoom out
ax2.plot(data6.loc[data6.LOCATION=='USA'].TIME, data6.loc[data6.LOCATION=='USA'].Value, color='blue')
ax2.set_title('UNITED STATES')


plt.savefig('revenue from Government')
```
